Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.


### 1. Probability Mass Function (PMF)

A Probability Mass Function (PMF) is used to describe the probability distribution of discrete random variables. It assigns a probability to each possible value that the discrete random variable can take. The sum of all the probabilities in a PMF is equal to 1.

#### Example: Rolling a Six-Sided Die

Consider the example of rolling a fair six-sided die. The random variable \(X\) represents the outcome of the die roll. The possible values of \(X\) are 1, 2, 3, 4, 5, and 6. Since the die is fair, each outcome has an equal probability of occurring.

The PMF for this scenario is:
\[ P(X = x) = \frac{1}{6} \]
for \(x = 1, 2, 3, 4, 5, 6\).

Here’s how the PMF looks in tabular form:

| \(X\) (Outcome) | \(P(X)\) (Probability) |
|-----------------|------------------------|
| 1               | \(\frac{1}{6}\)        |
| 2               | \(\frac{1}{6}\)        |
| 3               | \(\frac{1}{6}\)        |
| 4               | \(\frac{1}{6}\)        |
| 5               | \(\frac{1}{6}\)        |
| 6               | \(\frac{1}{6}\)        |

The PMF assigns a probability of \(\frac{1}{6}\) to each possible outcome of the die roll.

### 2. Probability Density Function (PDF)

A Probability Density Function (PDF) is used to describe the probability distribution of continuous random variables. Unlike the PMF, the PDF does not give probabilities directly. Instead, the area under the curve of the PDF over a specific interval represents the probability of the random variable falling within that interval. The total area under the PDF curve is equal to 1.

#### Example: Height Distribution of Adults

Consider the example of the height distribution of adult males in a population, which can be modeled as a continuous random variable. Suppose the heights are normally distributed with a mean (\(\mu\)) of 70 inches and a standard deviation (\(\sigma\)) of 3 inches.

The PDF for this normal distribution is given by:
\[ f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right) \]
where \( \mu = 70 \) and \( \sigma = 3 \).

Here’s the PDF formula for this specific normal distribution:
\[ f(x) = \frac{1}{\sqrt{2\pi(3^2)}} \exp\left(-\frac{(x - 70)^2}{2(3^2)}\right) \]

To find the probability that a randomly selected adult male has a height between 68 and 72 inches, we calculate the area under the PDF curve between these two values.

While the actual calculation involves integration, conceptually, the area under the PDF curve from \(x = 68\) to \(x = 72\) represents the probability of a height falling within this range.



Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?


### Cumulative Distribution Function (CDF)

A Cumulative Distribution Function (CDF) provides the probability that a random variable \(X\) will take a value less than or equal to a specific value \(x\). For a given value \(x\), the CDF is denoted as \(F(x)\) and is defined as:

\[ F(x) = P(X \leq x) \]

The CDF is useful because it provides a comprehensive way to describe the distribution of a random variable, offering insight into the probability of the variable falling within certain ranges. It applies to both discrete and continuous random variables.

### Example: CDF for Rolling a Six-Sided Die

Consider again the example of rolling a fair six-sided die. The random variable \(X\) represents the outcome of the die roll. The possible values of \(X\) are 1, 2, 3, 4, 5, and 6. Each outcome has an equal probability of \(\frac{1}{6}\).

The CDF for this discrete random variable can be constructed by summing the probabilities for each possible outcome up to \(x\):

\[ 
F(x) = \begin{cases} 
0 & \text{for } x < 1 \\
\frac{1}{6} & \text{for } 1 \leq x < 2 \\
\frac{2}{6} & \text{for } 2 \leq x < 3 \\
\frac{3}{6} & \text{for } 3 \leq x < 4 \\
\frac{4}{6} & \text{for } 4 \leq x < 5 \\
\frac{5}{6} & \text{for } 5 \leq x < 6 \\
1 & \text{for } x \geq 6 
\end{cases}
\]

Here’s the CDF in tabular form:

| \(X \leq x\) | \(F(x)\) (Cumulative Probability) |
|--------------|----------------------------------|
| \(x < 1\)    | 0                                |
| \(x < 2\)    | \(\frac{1}{6}\)                  |
| \(x < 3\)    | \(\frac{2}{6}\)                  |
| \(x < 4\)    | \(\frac{3}{6}\)                  |
| \(x < 5\)    | \(\frac{4}{6}\)                  |
| \(x < 6\)    | \(\frac{5}{6}\)                  |
| \(x \geq 6\) | 1                                |

### Example: CDF for Continuous Random Variable (Height Distribution)

Consider the example of the height distribution of adult males, which we previously modeled as a normal distribution with a mean (\(\mu\)) of 70 inches and a standard deviation (\(\sigma\)) of 3 inches.

The CDF for a continuous random variable \(X\) with a normal distribution is obtained by integrating the Probability Density Function (PDF) from \(-\infty\) to \(x\):

\[ F(x) = \int_{-\infty}^{x} f(t) \, dt \]

For the normal distribution, the CDF does not have a simple closed-form solution, but it is often denoted using the standard normal distribution and its corresponding CDF, \(\Phi(z)\), where \(z = \frac{x - \mu}{\sigma}\).

### Why CDF is Used

1. **Probability Calculation**: The CDF allows for easy calculation of the probability that a random variable falls within a certain range. For example, \(P(a \leq X \leq b) = F(b) - F(a)\).
  
2. **Descriptive Tool**: It provides a complete description of the distribution of the random variable, giving a cumulative perspective on how probabilities accumulate.
  
3. **Comparative Analysis**: It helps in comparing different random variables and their distributions by visualizing and analyzing their cumulative probabilities.

### Examples of Situations Where the Normal Distribution Might Be Used as a Model

1. **Heights of People**: The heights of adults of a specific gender and age group often follow a normal distribution. This is because height is influenced by many small, independent genetic and environmental factors.

2. **IQ Scores**: IQ scores are designed to follow a normal distribution with a mean of 100 and a standard deviation of 15. This is because the tests are standardized based on large populations to fit this distribution.

4. **Blood Pressure**: Systolic and diastolic blood pressure measurements in a healthy population typically follow a normal distribution, allowing for the identification of outliers and the establishment of normal ranges.

5. **Stock Returns**: While not perfect, the daily returns of a stock or the market index are often modeled using the normal distribution, especially for short time intervals. This helps in risk management and in the calculation of various financial metrics.

### Parameters of the Normal Distribution and Their Relation to the Shape of the Distribution

The normal distribution is characterized by two parameters: the mean (\(\mu\)) and the standard deviation (\(\sigma\)).

1. **Mean (\(\mu\))**:
   - The mean is the central point of the distribution and represents the average value of the data set.
   - The mean determines the location of the peak of the distribution. If the mean changes, the entire distribution shifts along the horizontal axis.
   - In the formula for the normal distribution, it appears as the center of the distribution:
     \[
     f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
     \]

2. **Standard Deviation (\(\sigma\))**:
   - The standard deviation measures the spread or dispersion of the distribution.
   - A smaller \(\sigma\) results in a steeper and narrower distribution, indicating that the data points are closely clustered around the mean.
   - A larger \(\sigma\) results in a flatter and wider distribution, indicating that the data points are more spread out from the mean.
   - The standard deviation affects the width of the bell curve and is present in both the scaling factor and the exponent of the normal distribution formula.
   



```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Different means
mu_values = [0, 2, 4]
sigma = 1
x = np.linspace(-5, 10, 1000)

for mu in mu_values:
    pdf = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, pdf, label=f'$\mu$={mu}, $\sigma$={sigma}')

plt.xlabel('x')
plt.ylabel('Density')
plt.title('Normal Distribution with Different Means')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Different standard deviations
mu = 0
sigma_values = [0.5, 1, 2]
x = np.linspace(-5, 5, 1000)

for sigma in sigma_values:
    pdf = stats.norm.pdf(x, mu, sigma)
    plt.plot(x, pdf, label=f'$\mu$={mu}, $\sigma$={sigma}')

plt.xlabel('x')
plt.ylabel('Density')
plt.title('Normal Distribution with Different Standard Deviations')
plt.legend()
plt.grid(True)
plt.show()
```

### Importance of Normal Distribution

The normal distribution is one of the most important probability distributions in statistics and is often referred to as the "Gaussian distribution" or "bell curve" due to its characteristic shape. Its importance stems from several key factors:

1. **Central Limit Theorem (CLT)**: The CLT states that the sum (or average) of a large number of independent, identically distributed random variables tends to follow a normal distribution, regardless of the original distribution of the variables. This makes the normal distribution a fundamental concept in probability and statistics.

2. **Predictability**: Many natural and human-made processes yield data that are approximately normally distributed. This predictability allows for easier analysis and modeling of real-world phenomena.

3. **Statistical Inference**: Many statistical methods and tests, such as t-tests, ANOVA, and regression analysis, assume that the underlying data follow a normal distribution. This assumption simplifies the mathematics and makes these techniques more robust and widely applicable.

4. **Symmetry and Measures**: The normal distribution is symmetric about its mean, making it easy to interpret measures of central tendency (mean, median, mode) and dispersion (standard deviation, variance).

### Real-Life Examples of Normal Distribution

1. **Heights of People**: In a given population, the heights of adults tend to follow a normal distribution. This distribution allows researchers to understand and predict height-related characteristics within the population.

   - Example: The average height of adult men in the U.S. is around 70 inches with a standard deviation of about 3 inches.

2. **IQ Scores**: IQ scores are designed to follow a normal distribution with a mean of 100 and a standard deviation of 15. This standardization allows for comparison and classification of cognitive abilities.

   - Example: About 68% of people have an IQ between 85 and 115, which is one standard deviation from the mean.

3. **Measurement Errors**: Errors in measurements often follow a normal distribution due to the accumulation of many small, random errors. This characteristic is crucial for quality control in manufacturing and scientific experiments.

   - Example: The length of machine-produced bolts may vary slightly due to machine precision, with the variations typically following a normal distribution.

4. **Blood Pressure**: Systolic and diastolic blood pressure measurements in a healthy population usually follow a normal distribution, enabling the establishment of normal ranges and identification of outliers.

   - Example: Normal systolic blood pressure for adults is typically around 120 mmHg with a standard deviation of about 10-20 mmHg.

5. **Grades in Exams**: Scores on standardized tests often follow a normal distribution. This distribution helps educators set grading scales and identify students' performance relative to their peers.

   - Example: The scores on the SAT, GRE, or other standardized tests are often normalized to follow a bell curve, allowing for consistent and fair comparisons.

### 1. What is Bernoulli Distribution?

The **Bernoulli distribution** is a discrete probability distribution for a random variable which takes the value 1 with probability \( p \) and the value 0 with probability \( 1 - p \). It is the simplest kind of distribution, named after the Swiss mathematician Jacob Bernoulli.

**Formula:**
\[ P(X = 1) = p \]
\[ P(X = 0) = 1 - p \]

Here:
- \( X \) is the random variable.
- \( p \) is the probability of success (the event happening).

**Example:**
Consider a coin toss where:
- Heads (success) = 1
- Tails (failure) = 0

If the coin is fair, the probability of getting heads is \( p = 0.5 \) and the probability of getting tails is \( 1 - p = 0.5 \).

So, for a fair coin:
- \( P(X = 1) = 0.5 \)
- \( P(X = 0) = 0.5 \)

- Represents the number of successes in a fixed number of independent Bernoulli trials.

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.


To find the probability that a randomly selected observation from a normally distributed dataset with a mean (\(\mu\)) of 50 and a standard deviation (\(\sigma\)) of 10 will be greater than 60, we can follow these steps:

1. **Calculate the Z-score** for the value 60.
2. **Use the Z-table** to find the probability corresponding to the calculated Z-score.
3. **Subtract the found probability** from 1 to get the probability of being greater than 60.

### Step 1: Calculate the Z-score

The Z-score is calculated using the formula:
\[ Z = \frac{X - \mu}{\sigma} \]

Where:
- \(X\) is the value for which we are calculating the Z-score (60 in this case).
- \(\mu\) is the mean of the dataset (50).
- \(\sigma\) is the standard deviation of the dataset (10).

Substituting the values:
\[ Z = \frac{60 - 50}{10} = \frac{10}{10} = 1 \]

### Step 2: Use the Z-table to find the probability

A Z-score of 1 corresponds to the cumulative probability up to that Z-score. We look up the Z-score of 1.0 in the Z-table (standard normal distribution table).

The Z-table gives the area to the left of the Z-score. For \( Z = 1 \), the area to the left is approximately 0.8413.

### Step 3: Subtract from 1

Since we want the probability of the observation being greater than 60, we need the area to the right of \( Z = 1 \):
\[ P(X > 60) = 1 - P(Z \leq 1) \]

Using the value from the Z-table:
\[ P(X > 60) = 1 - 0.8413 = 0.1587 \]

### Conclusion

The probability that a randomly selected observation from this dataset will be greater than 60 is approximately 0.1587, or 15.87%.

Q7: Explain uniform Distribution with an example

Uniform distribution is a type of probability distribution in which all outcomes are equally likely. Each variable within the range of the distribution has an equal probability of occurring. There are two main types of uniform distributions: discrete and continuous.

### Discrete Uniform Distribution

In a discrete uniform distribution, a finite number of values are equally likely to occur. For example, rolling a fair six-sided die results in a discrete uniform distribution because each of the six outcomes (1, 2, 3, 4, 5, 6) has an equal probability of \( \frac{1}{6} \).

### Continuous Uniform Distribution

In a continuous uniform distribution, all values within a certain range are equally likely to occur. The probability density function (pdf) of a continuous uniform distribution is constant, and the distribution is defined by the minimum value \(a\) and the maximum value \(b\).

The pdf for a continuous uniform distribution is given by:
\[ f(x) = \frac{1}{b-a} \]
for \( a \leq x \leq b \).

### Example of Continuous Uniform Distribution

Suppose a bus arrives at a bus stop every 10 minutes, and we arrive at the bus stop at a random time. We want to determine the probability of waiting less than 3 minutes for the bus.

Here, the waiting time \(X\) follows a uniform distribution between 0 and 10 minutes. Thus, \(a = 0\) and \(b = 10\).

The pdf of \(X\) is:
\[ f(x) = \frac{1}{10-0} = \frac{1}{10} = 0.1 \]
for \( 0 \leq x \leq 10 \).

To find the probability of waiting less than 3 minutes, we calculate the area under the pdf from \(x = 0\) to \(x = 3\):
\[ P(X < 3) = \int_{0}^{3} f(x) \, dx = \int_{0}^{3} 0.1 \, dx \]

Evaluating this integral:
\[ P(X < 3) = 0.1 \times (3 - 0) = 0.3 \]

So, the probability of waiting less than 3 minutes for the bus is 0.3, or 30%.

### Summary

- **Discrete Uniform Distribution:** All finite outcomes are equally likely (e.g., rolling a fair die).
- **Continuous Uniform Distribution:** All values within a given range are equally likely (e.g., waiting time for a bus that arrives uniformly every 10 minutes).

In both cases, the key characteristic of uniform distribution is that each possible outcome within the specified range is equally likely.

Q8: What is the z score? State the importance of the z score.


### What is the Z-score?

A Z-score, also known as a standard score, measures how many standard deviations a data point is from the mean of a distribution. It is calculated using the formula:

\[ Z = \frac{X - \mu}{\sigma} \]

Where:
- \( X \) is the value of the data point.
- \( \mu \) is the mean of the distribution.
- \( \sigma \) is the standard deviation of the distribution.

The Z-score indicates the position of a data point relative to the mean of the dataset.

### Importance of the Z-score

1. **Standardization**: Z-scores transform data from different scales to a common scale, allowing comparison across different datasets. This is useful when combining or comparing datasets with different units or magnitudes.

2. **Identifying Outliers**: Z-scores help identify outliers in data. Data points with Z-scores beyond ±3 are often considered outliers, as they lie far from the mean.

3. **Probability Calculation**: In a normal distribution, Z-scores are used to calculate the probability of a data point occurring within a certain range. This is done by referring to Z-tables, which provide the cumulative probability up to a given Z-score.

4. **Hypothesis Testing**: Z-scores are used in hypothesis testing, such as Z-tests, to determine whether to reject the null hypothesis. They allow researchers to determine if a sample mean significantly differs from the population mean.

5. **Comparison Across Different Distributions**: Z-scores enable comparison of scores from different distributions by standardizing them. For example, comparing exam scores from different tests with different means and standard deviations.

### Example Calculation

Consider a dataset with a mean (\(\mu\)) of 50 and a standard deviation (\(\sigma\)) of 10. To find the Z-score of a data point \( X = 70 \):

\[ Z = \frac{70 - 50}{10} = \frac{20}{10} = 2 \]

This Z-score of 2 means that the data point 70 is 2 standard deviations above the mean.

### Summary

Z-scores are crucial in statistics for standardizing data, identifying outliers, calculating probabilities, conducting hypothesis tests, and enabling comparisons across different distributions. By providing a measure of how far a data point is from the mean, Z-scores offer a standardized way to interpret and analyze data.

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem

### What is the Central Limit Theorem?

The Central Limit Theorem (CLT) is a fundamental statistical principle that states that the sampling distribution of the sample mean (or sum) of a sufficiently large number of independent, identically distributed (i.i.d.) random variables approaches a normal distribution, regardless of the original distribution of the population from which the samples are drawn. The theorem is formally stated as:

For a population with any shape distribution, the sampling distribution of the sample mean will approach a normal distribution as the sample size \( n \) increases, provided \( n \) is sufficiently large (typically \( n \geq 30 \) is considered sufficient).

Mathematically, if \( X_1, X_2, \ldots, X_n \) are i.i.d. random variables with mean \( \mu \) and standard deviation \( \sigma \), then the sample mean \( \bar{X} \) is given by:

\[ \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i \]

As \( n \) becomes large, the distribution of \( \bar{X} \) approaches a normal distribution with mean \( \mu \) and standard deviation \( \frac{\sigma}{\sqrt{n}} \):

\[ \bar{X} \sim N\left(\mu, \frac{\sigma}{\sqrt{n}}\right) \]

### Significance of the Central Limit Theorem

1. **Foundation for Inferential Statistics**: The CLT provides the theoretical foundation for making inferences about population parameters from sample statistics. It justifies the use of the normal distribution in hypothesis testing and confidence interval estimation, even when the population distribution is not normal.

2. **Simplification of Analysis**: Because of the CLT, many statistical methods and procedures can be applied under the assumption of normality, which simplifies analysis. This is particularly useful when dealing with complex or unknown population distributions.

3. **Enables Standardization**: The CLT allows for the standardization of sample means, making it possible to use Z-scores and other standard normal distribution tools to determine probabilities and critical values.

4. **Applicability Across Various Fields**: The CLT is widely applicable in different fields such as economics, engineering, psychology, and natural sciences, where it is often impractical to know the exact distribution of the population.

5. **Large Sample Theory**: The CLT is the cornerstone of large sample theory, enabling the use of large sample approximations to make statistical decisions and derive properties of estimators.

### Example of Central Limit Theorem

Consider a population with an unknown distribution having mean \( \mu = 50 \) and standard deviation \( \sigma = 10 \). If we take random samples of size \( n = 40 \) from this population and calculate the mean of each sample, the distribution of these sample means will be approximately normal with:

- Mean \( \mu = 50 \)
- Standard deviation \( \frac{\sigma}{\sqrt{n}} = \frac{10}{\sqrt{40}} \approx 1.58 \)

Even if the original population distribution is skewed or has outliers, the distribution of the sample means will approximate a normal distribution as the sample size increases.

### Summary

The Central Limit Theorem is a key concept in statistics that explains why the normal distribution is so prevalent in statistical analysis. It states that the distribution of sample means approximates a normal distribution as the sample size grows, regardless of the population's distribution. This theorem is crucial for the validity of many statistical methods and underpins much of inferential statistics, making it a cornerstone of statistical theory and practice.

Q10: State the assumptions of the Central Limit Theorem.

The Central Limit Theorem (CLT) is a powerful statistical concept, but it relies on certain assumptions to hold true. Here are the key assumptions of the CLT:

1. **Independence**: The random variables (samples) must be independent of each other. This means that the selection of one sample should not influence the selection of another.

2. **Identically Distributed**: The random variables should be identically distributed, meaning they should all come from the same population with the same probability distribution. Each sample should have the same mean and variance.

3. **Sample Size**: The sample size should be sufficiently large. While there is no hard and fast rule for what constitutes a "large" sample size, a common guideline is \( n \geq 30 \). However, the required sample size can be smaller if the population distribution is already approximately normal.

### Additional Considerations

- **Finite Variance**: The population from which samples are drawn should have a finite variance. If the population variance is infinite, the CLT does not apply.
- **Underlying Distribution Shape**: While the CLT states that the distribution of the sample mean will approach normality, the rate of this convergence can depend on the shape of the underlying population distribution. For populations with heavy tails or significant skewness, larger sample sizes may be necessary.

### Summary

The assumptions of the Central Limit Theorem can be summarized as follows:

1. **Independence**: Samples must be independent.
2. **Identically Distributed**: Samples must be identically distributed.
3. **Large Sample Size**: A sufficiently large sample size is required, typically \( n \geq 30 \).
4. **Finite Variance**: The population should have a finite variance.

These assumptions ensure that the sampling distribution of the sample mean will be approximately normal, allowing for the application of various statistical methods and inferences based on the normal distribution.
